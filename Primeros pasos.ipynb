**Q (Husimi) Function and Wigner functions in coherent and number states**

```python
beta = 1+1j
xvec = np.arange(-5, 5, 1e-1)
yvec = np.arange(-5, 5, 1e-1)
Qnumber=np.zeros([len(xvec),len(yvec)],float)
Qcoherent=np.zeros([len(xvec),len(yvec)],float)
for i in range(len(xvec)):
    for j in range(len(yvec)):
       Qcoherent[i,j]=1/np.pi*np.exp(-np.abs(xvec[i]+1j*yvec[j]-beta)**2)
       Qnumber[i,j]=1/np.pi*np.exp(-np.abs(xvec[i]+1j*yvec[j])**2)*np.abs(xvec[i]+1j*yvec[j])**6/6
X, Y = np.meshgrid(xvec, yvec)
```

**THE COHERENT STATE**
$$
Q_c(\alpha)=\frac{1}{\pi}|<\alpha|\beta>|^2=\frac{1}{\pi}exp(-|\alpha-\beta|^2)
$$
**THE NUMBER STATE**
$$
Q_n(\alpha)=\frac{1}{\pi}|<\alpha|\beta>|^2=\frac{1}{\pi}exp(-|\alpha|^2)|\frac{|\alpha|^{2n}}{n!}
$$

```python
fig = plt.figure(figsize=(8, 6))
plt.contourf(X, Y, np.transpose(Qnumber), 50, cmap=mpl.colormaps['inferno'])
plt.colorbar();
```

```python
fig = plt.figure(figsize=(8, 6))
plt.contourf(X, Y, np.transpose(Qcoherent), 50, cmap=mpl.colormaps['inferno'])
plt.colorbar();
```
